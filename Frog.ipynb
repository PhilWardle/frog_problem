```python
def frogStandUp(final_dist=100,max=6):
    
    # I wrote this quickly in a pub
    # Don't judge me
    
    total_dist = 1
    
    while total_dist <= final_dist:
        
        import random
        
        cap = 10**max
        
        trials = 0
        total_leaps = 0
        
        while trials <= cap:
            
            dist = total_dist
            leaps = 0
            
            while dist > 0:
                this_jump = int(random.random()*(dist)+1)
                dist -= this_jump
                leaps += 1
                
            total_leaps += leaps
            
            trials += 1
            
        print("{0}\t{1}".format(total_dist,(total_leaps*1.0)/trials))
    
        total_dist += 1
    
    return "DONE"
```

```python
def frog(distance=100,trials=1000):
    import random
    trial = 0
    totalJumps = 0
    while trial <= trials:
        trial += 1
        
        jumps = 0
        distanceCovered = 0
        
        while distanceCovered < distance:
            distanceCovered += random.randint(1,distance-distanceCovered)
            jumps += 1
            
        totalJumps += jumps
        
        print(f'Trial {trial}, jumps {jumps}, running average {totalJumps/trial:5.2f}')
    print('Done')
```

```python
frog()
```

```python
frog(10,10)
```

Same code but write to file

```python
def frogWrite(distance=100,trials=1000):
    outfile = open('frog.txt','w')
    import random
    trial = 0
    totalJumps = 0
    while trial <= trials:
        trial += 1
        
        jumps = 0
        distanceCovered = 0
        
        while distanceCovered < distance:
            distanceCovered += random.randint(1,distance-distanceCovered)
            jumps += 1
            
        totalJumps += jumps
        
        print(f'Trial {trial}, jumps {jumps}, running average {totalJumps/trial:5.2f}')
        outfile.write(f'Trial: {trial}, jumps: {jumps}, running average: {totalJumps/trial:5.2f}\n')
    print('Done')
```

```python
frogWrite()
```
